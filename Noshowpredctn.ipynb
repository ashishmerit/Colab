```python
import pandas as pd
import random
from datetime import datetime, timedelta

doctors_df = pd.read_csv("doctors_dataset_100.csv")
patients_df = pd.read_csv("patients_1000.csv")

if "doctor_id" not in doctors_df.columns:
    doctors_df["doctor_id"] = ["D" + str(i+1).zfill(3) for i in range(len(doctors_df))]

time_slots = ["09:00–09:30", "10:00–10:30", "11:00–11:30", "12:00–12:30",
              "02:00–02:30", "03:00–03:30", "04:00–04:30", "05:00–05:30"]

appointments = []

for doctor in doctors_df.itertuples():
    for day_offset in range(30):  # 30 days
        date_obj = datetime.today() + timedelta(days=day_offset)
        date = date_obj.strftime("%Y-%m-%d")
        day_of_week = date_obj.strftime("%A")

        num_appointments = random.randint(3, 5)  # 3 to 5 per doctor per day
        for _ in range(num_appointments):
            patient = patients_df.sample(1).iloc[0]
            appointment_id = f"APT{len(appointments)+1:05d}"
            appointments.append({
                "appointment_id": appointment_id,
                "doctor_id": doctor.doctor_id,
                "patient_id": patient.patient_id,
                "date": date,
                "day_of_week": day_of_week,
                "time_slot": random.choice(time_slots),
                "status": random.choices(["completed", "no-show"], weights=[0.8, 0.2])[0]
            })

appointments_df = pd.DataFrame(appointments)
appointments_df.to_csv("appointments_12000.csv", index=False)

print("Generated appointments_12000.csv")
appointments_df.head()
```

```python
import pandas as pd
import random
from datetime import timedelta

appointments = pd.read_csv("appointments_updated.csv")
patients = pd.read_csv("patients_1000.csv")


appointments = pd.merge(appointments, patients[["patient_id", "age", "gender"]], on="patient_id", how="left")


appointments["date"] = pd.to_datetime(appointments["date"])


appointments["booking_date"] = appointments["date"].apply(
    lambda x: x - timedelta(days=random.randint(1, 20))
)


appointments["days_waiting"] = (appointments["date"] - appointments["booking_date"]).dt.days

appointments.sort_values(by=["patient_id", "date"], inplace=True)
appointments["is_first_visit"] = ~appointments.duplicated(subset=["patient_id"], keep="first")


appointments["reminder_sent"] = [random.choice([True] * 8 + [False] * 2) for _ in range(len(appointments))]


appointments["has_chronic_condition"] = [random.choice([True] * 3 + [False] * 7) for _ in range(len(appointments))]


def get_age_group(age):
    if age < 13:
        return "Child"
    elif age < 20:
        return "Teen"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

appointments["age_group"] = appointments["age"].apply(get_age_group)



appointments["no_show"] = appointments["status"].apply(lambda x: 1 if x == "no-show" else 0)

appointments.to_csv("appointments_final.csv", index=False)

print("appointments_final.csv created with merged patient info and all required features.")
appointments.head()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

# Define the Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Define the parameter grid with fewer options for faster execution
param_dist = {
    'n_estimators': [100, 150, 200],            # Number of trees in the forest
    'max_depth': [None, 10, 20],                 # Maximum depth of the tree
    'min_samples_split': [2, 5],                  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2],                   # Minimum number of samples required at a leaf node
    'bootstrap': [True, False]                    # Whether to use bootstrap samples
}

# Perform random search with 3-fold cross-validation, n_iter defines the number of combinations to test
random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=param_dist,
                                   n_iter=10, cv=3, n_jobs=-1, verbose=2, scoring='accuracy', random_state=42)

# Fit the model to the training data
random_search.fit(X_train_clean, y_train)

# Get the best parameters from the random search
best_params = random_search.best_params_
print("Best parameters found: ", best_params)

# Train the Random Forest model with the best parameters
best_rf_model = random_search.best_estimator_

# Predict using the best model
y_pred = best_rf_model.predict(X_test_clean)

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Optional: Plot ROC Curve for the best model
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_probs = best_rf_model.predict_proba(X_test_clean)[:, 1]  # Probabilities for class 1
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
auc_score = roc_auc_score(y_test, y_probs)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"AUC = {auc_score:.2f}")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest (RandomizedSearchCV)")
plt.legend()
plt.grid(True)
plt.show()
```
